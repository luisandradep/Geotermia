Para el cálculo de la temperatura en la base de la corteza superior (granitos viejos), $T_1$, contamos con lo siguiente:

La profundidad inicial es $Z_0 = 0 m$. Mientras que la profundidad final es $Z_1= 8000 m$.

La temperatura en superficie $T_0$ es de 10°C.

Como dato tenemos que el gradiente geotérmico de la corteza superior es: $(\frac{\Delta T}{\Delta Z})_{sup} = 1,9 x 10^{-2} \frac{°C}{m}$

```python
Z0 = 0
T0 = 10
Z1 = 8000
GradTerm1 = 1.9*10**-2
```

Al desglosar la ecuación del gradiente geotérmico tenemos que:
\begin{equation}(\frac{\Delta T}{\Delta Z})_{sup} = \frac{T_1 - T_0}{Z_1 - Z_0}\end{equation}

Despejamos la ecuación para hallar $T_1$ dado que contamos con los demás datos:
\begin{equation}T_1 = (\frac{\Delta T}{\Delta Z})_{sup} * (Z_1 - Z_0) + T_0\end{equation}
\begin{equation}T_1 = 1,9 x 10^{-2}\frac{°C}{m} * (8000m - 0m) + 10°C \end{equation}

```python
T1= GradTerm1 * (Z1-Z0) + T0
print("T1 = " + str(T1) + " °C")
```

La temperatura basal de la corteza $T_1$ es de **162 °C**. Esta temperatura a su vez es la del tope de la corteza inferior

La relación entre el flujo de calor $q$ y la generación de calor $A$ en una capa se rige por la siguiente relación:
\begin{equation} q_{inf} - q_{sup} = -A * (Z_{inf} - Z_{sup})\end{equation}
Donde $q_{inf}$ corresponde al flujo de calor en la base de la capa y $q_{sup}$ en el tope.

$Z_{inf}$ es la profundidad de base de la capa y $Z_{sup}$ la profundidad de su tope.

Podemos estimar el flujo de calor en la base de la corteza inferior $q_2$ adaptando y despejando la ecuación anterior:
\begin{equation} q_{2} = q_{1} - A_{c-inf} * (Z_{2} - Z_{1})\end{equation}
\begin{equation} q_{2} = 3,2*10^{-2} \frac{W}{m^2}- 0,25*10^{-6}\frac{W}{m^3} * (35000 - 8000)m\end{equation}

```python
q1 = 3.2*10**-2
Ainf = 0.25*10**-6
Z2 = 35000
Z1 = 8000
q2 = q1 - Ainf * (Z2 - Z1)
print("q2 = " + str(q2)+ " W/m^2")
```

Como resultado obtenemos que el flujo de calor en la base de la corteza inferior **$\bf{q_2}$** es de $\bf{2,525 * 10^{-2} \frac{W}{m^2}}$

Una vez conocidos los flujos de calor de la corteza inferior, podemos estimar el correspondiente gradiente geotérmico determinado por la siguiente ecuación:

\begin{equation} q = k*\frac{\Delta T}{\Delta Z}\end{equation}
Despejamos el gradiente y obtenemos la siguiente expresión:
\begin{equation} \frac{\Delta T}{\Delta Z} = \frac{q}{k}\end{equation}

El valor q lo podemos obtener promediando los valores tope $q_1$ y base $q_2$:
\begin{equation}q = \frac{q_1 + q_2}{2}\end{equation}
Además tenemos la conductividad calórica de la corteza inferior $k_1$ es de 3 $\frac{W}{m°C*}$

```python
q = (q1 + q2) / 2
k1 = 3
GradTerm2 = round(q / k1, 4)        #redondear a 4 decimales
print("Preomedio de flujo de calor en corteza inferior = " + str(q) + " W/m^2")
print("Grad. geoterm. cort_inf = " + str(GradTerm2) + " °C/m)")
```

El gradiente geotérmico de la corteza inferior es de $\bf{9,5*10^{-3}} \frac{°C}{m}$

Finalmente con los datos obtenidos calculamos la temperatura basal de la corteza inferior $T_2$ a partir de la siguiente relación:
\begin{equation}(\frac{\Delta T}{\Delta Z})_{inf} = \frac{T_2 - T_1}{Z_2 - Z_1}\end{equation}
Que al ser despejada nos queda de la siguiente forma, donde conocemos las demás variables:
\begin{equation}T_2 = T_1 + (Z_2 - Z_1)*(\frac{\Delta T}{\Delta Z})_{inf}\end{equation}

```python
T2 = T1 + (Z2 - Z1)*GradTerm2
print("T2 = " + str(T2) + " °C")
```

La temperatura basal resultante de la corteza terrestre $T_2$ es de **418,5 °C**

Con este resultado podemos resumir los valores estimados para las capas inferiores a la superficie desconocidos:

Temperatura basal de corteza superior $\bf{T_1 = 162 °C}$

Flujo de calor basal de la corteza inferior $\bf{q_2 = 2,525 * 10^{-2} \frac{W}{m^2}}$

Gradiente geotérmico de la corteza inferior $\bf{(\frac{\Delta T}{\Delta Z})_{inf} = 9,5*10^{-3} \frac{°C}{m}}$

Temperatura basal de la corteza inferior $\bf{T_2 = 418,5 °C}$
